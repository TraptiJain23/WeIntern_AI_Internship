##9)  Standardize inconsistent categorical labels

```python
# All categorical columns
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    print(f"\n===== {col} =====")
    print(df[col].value_counts(dropna=False))
```

### Observation

Several categorical columns contained inconsistent labels.

For example:
- M and Male
- F and Female
- Y and Yes

These values must be standardized to ensure consistency during analysis.

```python
# Standardize Gender
df["Gender"] = df["Gender"].replace({
    "male": "Male",
    "M": "Male",
    "female": "Female",
    "F": "Female"
})

# Standardize Parental involvement
df["Parental involvement"] = df["Parental involvement"].replace({
    "high": "High",
    "medium": "Medium",
    "low": "Low"
})

# Standardize Internet Access
df["Internet Access"] = df["Internet Access"].replace({
    "yes": "Yes",
    "Y": "Yes",
    "no": "No",
    "N": "No"
})

# Standardize Access_to_Resources
df["Access_to_Resources"] = df["Access_to_Resources"].replace({
    "high": "High",
    "med": "Medium",
    "low": "Low"
})

# Standardize Extracurricular_Activities
df["Extracurricular_Activities"] = df["Extracurricular_Activities"].replace({
    "yes": "Yes",
    "Y": "Yes",
    "no": "No",
    "N": "No"
})

# Standardize Motivation_Level
df["Motivation_Level"] = df["Motivation_Level"].replace({
    "high": "High",
    "medium": "Medium",
    "low": "Low"
})

# Standardize Teacher_Quality
df["Teacher_Quality"] = df["Teacher_Quality"].replace({
    "good": "Good",
    "avg": "Average"
})

# Standardize Peer_Influence
df["Peer_Influence"] = df["Peer_Influence"].replace({
    "positive": "Positive",
    "neutral": "Neutral",
    "negative": "Negative"
})

# Standardize Learning_Disabilities
df["Learning_Disabilities"] = df["Learning_Disabilities"].replace({
    "yes": "Yes",
    "Y": "Yes",
    "no": "No",
    "N": "No"
})

# Standardize Parental_Education_Level
df["Parental_Education_Level"] = df["Parental_Education_Level"].replace({
    "bachelor": "Bachelor",
    "masters": "Master"
})

# Standardize Student Id
df["Student ID"] = [
    f"S{str(i+1).zfill(5)}"
    for i in range(len(df))
]

# Remove extra spaces from Competency Areas
df["Competency Areas"] = df["Competency Areas"].str.strip()
```

###Verification

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"\n===== {col} =====")
    print(df[col].value_counts(dropna=False))
```

##10) Rename unclear columns

```python
df.rename(columns={"Notes":"Remarks"}, inplace=True)
```

## 11) Data Type Conversion


```python
# Convert Age
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

# Convert Attendance (remove % first)
df["Attendance"] = (
    df["Attendance"]
    .astype(str)
    .str.replace("%", "", regex=False)
)

df["Attendance"] = pd.to_numeric(df["Attendance"], errors="coerce")

# Convert other numerical columns
num_cols = [
    "Hours_Studied",
    "Assignment Score",
    "Sleep_Hours",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Physical_Activity",
    "Distance_from_Home"
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

### Check updated data types

```python
df.info()
```

## 12) Missing Value Handling

```python
# Numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Categorical columns
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values handled successfully.")
```

```python
print("Total Missing Values:", df.isnull().sum().sum())
```



Missing values were successfully handled using median imputation for numerical features and mode imputation for categorical features.



After the imputation process, no missing values remained in the dataset, making it suitable for further analysis and modeling.


### Justification

Median imputation was used for numerical features because it is less affected by extreme values.

Mode imputation was used for categorical features because it preserves the most frequently occurring category.
