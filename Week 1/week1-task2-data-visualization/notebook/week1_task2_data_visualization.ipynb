# 4) Visualization 1: Distribution Analysis

## Histogram of Final Exam Scores
Purpose:
- Understand score distribution.
- Check whether scores are balanced, skewed, or clustered.

This visualization examines how students' final exam scores are distributed across the dataset.


```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["Final Exam Score"],
    bins=20,
    kde=True
)

plt.title("Distribution of Final Exam Scores")
plt.xlabel("Final Exam Score")
plt.ylabel("Frequency")

plt.savefig("visuals/visualization_1_distribution_analysis.png", bbox_inches="tight")
plt.show()
```

### Observation & Insights

**What the graph displays:**
This histogram shows the distribution of students' final exam scores across the dataset. The bars represent the number of students within different score ranges.

**Visible pattern:**
The KDE curve remains relatively flat throughout most of the distribution.

**Strength of the pattern:**
The pattern is weak and does not indicate a clear peak, skewness, or clustering.

**Possible reason:**
The dataset contains students with varying performance levels, resulting in a wide spread of scores.

**Key Findings:**
- Final exam scores range from approximately 30 to 100.
- No major outliers are visible in the distribution.
- The score distribution appears balanced rather than heavily skewed.
- There is no strong evidence of positive or negative skewness.
- Student performance appears diverse, with representation across low, medium, and high score groups.

# 11) Visualization 8: Correlation Analysis
## Correlation Heatmap of Numerical Features

Purpose:

- Identify relationships among numerical variables.
- Understand factors associated with student performance.

This visualization examines the strength and direction of relationships between numerical features in the dataset.

```python
# ==========================================
# Correlation Matrix
# ==========================================

numerical_cols = df.select_dtypes(include=['int64','float64'])

corr_matrix = numerical_cols.corr()

corr_matrix
```

### Matrix Analysis

The correlation matrix confirms the observations from the heatmap. The strongest positive correlation is between Previous Scores and Final Exam Score (0.0276), which is still very weak. Overall, no numerical feature shows a strong linear relationship with student performance.

```python
# ==========================================
# Heatmap
# ==========================================

plt.figure(figsize=(12,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)

plt.title("Correlation Heatmap of Numerical Features")

plt.savefig("visuals/visualization_8_correlation_heatmap.png", bbox_inches="tight")
plt.show()
```

```python
# ==========================================
# Strongest Correlations
# ==========================================

corr_pairs = corr_matrix.unstack()

corr_pairs = corr_pairs[
    corr_pairs.index.get_level_values(0)
    != corr_pairs.index.get_level_values(1)
]

corr_pairs = corr_pairs.sort_values(ascending=False)

print("Top Positive Correlations:")
print(corr_pairs.head(10))
```

### Observation & Insights

**What the graph displays:** The correlation heatmap shows the strength and direction of relationships among numerical features in the dataset.

**Visible pattern:** Most correlation values are very close to zero, indicating weak relationships between variables.

**Strength of the pattern:** The overall correlations are extremely weak, with no strong positive or negative relationships visible.

**Possible reason:** The variables appear to be largely independent of one another, suggesting that student performance is not strongly influenced by any single numerical factor in this dataset.

**Key findings:**

- No strong correlations are present among the numerical features.
- Previous Scores show the highest positive correlation with Final Exam Score, but the relationship remains very weak.
- Hours Studied and Attendance have little to no correlation with Final Exam Score.
- Most correlation coefficients lie close to zero, indicating weak feature interactions.
- The dataset does not exhibit strong linear relationships between variables.
